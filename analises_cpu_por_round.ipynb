```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd
import numpy as np, scipy.stats as st
```

```python
#l_data   = pd.read_csv("./all/large_Monitor.csv")
#l_data  = pd.read_csv("./all/largex_Monitor.csv")
#siteB_data = pd.read_csv("TxMonitor-siteB_2xlarge.csv")
siteB_data = pd.read_csv("2xlarge_Monitor.csv")
#l_round   = pd.read_csv("./all/large.csv")
#l_round  = pd.read_csv("./all/largex.csv")
#l_round = pd.read_csv("./all/largex2.csv")
```

```python
print(siteB_data['cpu(%)'])
```

```python
siteB__cpu_mean = []
siteB__mem_mean = []
siteB__cpu_erro = []
siteB_filtro = []
for i in range(1,15):
    l_round_x = siteB_data.query('round==%s'%i)
    cpu = l_round_x['cpu(%)']
    memoria = l_round_x['memoria(%)']
    siteB_filtro.append(cpu.tolist())
    #print(cpu)
    cpu_mean = np.mean(cpu)
    mem_mean = np.mean(memoria)
    cpu_erro = st.t.interval(0.05, len(cpu)-1, loc=np.mean(cpu), scale=st.sem(cpu))
    siteB__cpu_mean.append(cpu_mean)
    siteB__mem_mean.append(mem_mean)
    siteB__cpu_erro.append(cpu_erro[1] - cpu_erro[0])
```

```python
print(siteB__cpu_mean)
print(siteB__cpu_erro)
print(siteB_filtro)
```
