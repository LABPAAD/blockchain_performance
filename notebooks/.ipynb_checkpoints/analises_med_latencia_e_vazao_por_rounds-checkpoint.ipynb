```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
def formatarq(txt,nm):
  with open(txt, 'r') as fd:
    txt = fd.read()  # Ler todo o arquivo
  txt = txt.replace('|',',')
  txt = txt.replace(' ','')
  with open(nm, 'w') as fd:
    fd.write(txt)
```

```python
formatarq('exp_cluster1.txt','cluster1.txt')
formatarq('exp_cluster2.txt','cluster2.txt')
```

```python
cl1 = pd.read_csv('cluster1.txt')
cl2 = pd.read_csv('cluster2.txt')
```

```python
print(cl1['AvgLatency(s)'])
```

```python
print(cl2['AvgLatency(s)'])
```

```python
def media(cl,var):
  mdlt = []
  rdns = cl[var]
  partes = [rdns[index : index + 10] for index in range(0, len(rdns), 10)]
  for i in partes:
    mdlt.append(np.mean(i))
  return mdlt
```

```python
mdlt1 = media(cl1,'AvgLatency(s)')
mdlt2 = media(cl2,'AvgLatency(s)')
```

```python
print(mdlt1)
print(mdlt2)
```

```python
x = [10, 20, 40, 60,80, 100, 120,140, 160,180, 200]

fig,ax = plt.subplots(figsize =(15,8))
ax.set_xlabel('Variação de (tps) por round', fontsize=20)
ax.set_ylabel('Média de Latência', fontsize=20)

ax.plot(mdlt1, marker="*", label="Cluster 1", linewidth=4, markersize=20)
ax.plot(mdlt2, marker=".", label="Cluster 2", linewidth=4, markersize=21)
ax.set_xticks(range(12))
ax.set_xticklabels(x, fontsize=18)
plt.yticks([0,2,4,6,8,10,12], fontsize=18)
plt.legend()
plt.savefig('analise_medlat_rouds.pdf')
plt.show()
```

```python
mdvz1 = media(cl1,'Throughput(TPS)')
mdvz2 = media(cl2,'Throughput(TPS)')
print(mdvz1)
print(mdvz2)
```

```python
x = [10, 20, 40, 60,80, 100, 120,140, 160,180, 200]

fig,ax = plt.subplots(figsize =(15,8))
#ax.set_title('Ethereum Provider - t2.2xlarge')
ax.set_xlabel('Variação de (tps) por round', fontsize=20)
ax.set_ylabel('Média da Vazão', fontsize=20)

#ax.grid(linestyle='dotted')
ax.plot(mdvz1, marker="*", label="Cluster 1", linewidth=4, markersize=20)
ax.plot(mdvz2, marker=".", label="Cluster 2", linewidth=4, markersize=21)
ax.set_xticks(range(12))
ax.set_xticklabels(x, fontsize=18)
plt.yticks([0,5,10,15,20,25,30,35,40,45,50], fontsize=18)
plt.legend()
plt.savefig('analise_medvaz_rouds.pdf')
plt.show()
```

```python
def erroMedia(cl,var):
  erro = []
  rdns = cl[var]
  partes = [rdns[index : index + 10] for index in range(0, len(rdns), 10)]
  for i in partes:
    interval = st.t.interval(alpha=0.90,df=len(i)-1,loc=np.mean(i),scale=st.sem(i))
    erro.append(np.mean(i) - interval[0])
  return erro
```

```python
print(erroMedia(cl1,'AvgLatency(s)'))
```

```python
x = [10, 20, 40, 60,80, 100, 120,140, 160,180, 200]
#yerr1 = [aux if aux > 0 else aux*-1 for aux in [x[id]-mdlt1[id] for id in range(0, 11)]]
yerro1 = erroMedia(cl1,'AvgLatency(s)')
print(yerro1)
#yerr2 = [aux if aux > 0 else aux*-1 for aux in [x[id]-mdlt2[id] for id in range(0, 11)]]
yerro2 = erroMedia(cl2,'AvgLatency(s)')
fig = plt.figure()
plt.subplots(figsize =(15,8))
plt.xlabel('Variação de (tps) por round', fontsize=20)
plt.ylabel('Média de Latência',size=20)
#yerr = np.linspace(0.05, 0.2, 10)

plt.errorbar(x, mdlt1, yerr=yerro1,marker='*',label="Cluster 1", linewidth=4, markersize=20)

plt.errorbar(x, mdlt2, yerr=yerro2,marker='.',label="Cluster 2", linewidth=4, markersize=20)

plt.xticks(x, fontsize=18)
#plt.yticks([-150,-100,-50,0,50,100,150,200], fontsize=18)
plt.legend()
plt.savefig('analise_medlat_erro_rouds.pdf')
```

```python
x = [10, 20, 40, 60,80, 100, 120,140, 160,180, 200]
#yerr1 = [aux if aux > 0 else aux*-1 for aux in [x[id]-mdvz1[id] for id in range(0, 11)]]
yerro1 = erroMedia(cl1,'Throughput(TPS)')
print(yerro1)
#yerr2 = [aux if aux > 0 else aux*-1 for aux in [x[id]-mdvz2[id] for id in range(0, 11)]]
yerro2 = erroMedia(cl2,'Throughput(TPS)')
fig = plt.figure()
plt.subplots(figsize =(15,8))
plt.xlabel('Variação de (tps) por round', fontsize=20)
plt.ylabel('Média da Vazão',size=20)
#yerr = np.linspace(0.05, 0.2, 10)

plt.errorbar(x, mdvz1, yerr=yerro1,marker='*',label="Cluster 1", linewidth=4, markersize=20)

plt.errorbar(x, mdvz2, yerr=yerro2,marker='.',label="Cluster 2", linewidth=4, markersize=20)

plt.xticks(x, fontsize=18)
#plt.yticks([-150,-100,-50,0,50,100,150,200], fontsize=18)
plt.legend()
plt.savefig('analise_medvaz_erro_rouds.pdf')
```
