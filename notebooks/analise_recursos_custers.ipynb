```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd
import numpy as np, scipy.stats as st
```

```python
vm1 = pd.read_csv("TxMonitor-vm1-cluster1.csv")
vm2 = pd.read_csv("TxMonitor-vm2-cluster1.csv")
vm3 = pd.read_csv("TxMonitor-vm3-cluster1.csv")
vm4 = pd.read_csv("TxMonitor-vm4-cluster1.csv")
```

```python
vm1 = pd.read_csv("TxMonitor-vm1-cluster2.csv")
vm2 = pd.read_csv("TxMonitor-vm2-cluster2.csv")
vm3 = pd.read_csv("TxMonitor-vm3-cluster2.csv")
```

```python
host1__cpu_mean = []
host1__mem_mean = []
host1__cpu_erro = []
host1__mem_erro = []
host1__neti_mean = []
host1__neto_mean = []
host1__neto = []
host1_neti = []
host1__neti_erro = []
host1__neto_erro = []
host1_filtro = []

host2__cpu_mean = []
host2__mem_mean = []
host2__cpu_erro = []
host2__mem_erro = []
host2__neti_mean = []
host2__neto_mean = []
host2__neto = []
host2_neti = []
host2__neti_erro = []
host2__neto_erro = []
host2_filtro = []

host3__cpu_mean = []
host3__mem_mean = []
host3__cpu_erro = []
host3__mem_erro = []
host3__neti_mean = []
host3__neto_mean = []
host3__neto = []
host3_neti = []
host3__neti_erro = []
host3__neto_erro = []
host3_filtro = []
'''
host4__cpu_mean = []
host4__mem_mean = []
host4__cpu_erro = []
host4__mem_erro = []
host4__neti_mean = []
host4__neto_mean = []
host4__neto = []
host4_neti = []
host4__neti_erro = []
host4__neto_erro = []
host4_filtro = []
'''
```

```python
for i in range(1,12):
#host1  
    l_round_1 = vm1.query('round==%s'%i)
    cpu1 = l_round_1['cpu(%)']
    memoria1 = l_round_1['memoria(%)']
    host1_filtro.append(cpu1.tolist())
    cpu_mean1 = np.mean(cpu1)
    mem_mean1 = np.mean(memoria1)

    neti1 = l_round_1['Net-in(bytes)']
    neto1 = l_round_1['Net-out(bytes)']
    neti_mean1 = np.mean(neti1)
    neto_mean1 = np.mean(neto1)

    cpu_erro1 = st.t.interval(0.05, len(cpu1)-1, loc=np.mean(cpu1), scale=st.sem(cpu1))
    host1__cpu_erro.append(cpu_erro1[1] - cpu_erro1[0])

    mem_erro1 = st.t.interval(0.05, len(memoria1)-1, loc=np.mean(memoria1), scale=st.sem(memoria1))
    host1__mem_erro.append(mem_erro1[1] - mem_erro1[0])

    neti_erro1 = st.t.interval(0.05, len(neti1)-1, loc=np.mean(neti1), scale=st.sem(neti1))
    host1__neti_erro.append(neti_erro1[1] - neti_erro1[0])

    neto_erro1 = st.t.interval(0.05, len(neto1)-1, loc=np.mean(neto1), scale=st.sem(neto1))
    host1__neto_erro.append(neto_erro1[1] - neto_erro1[0])

    host1__cpu_mean.append(cpu_mean1)
    host1__mem_mean.append(mem_mean1)
    host1__neti_mean.append(neti_mean1)
    host1__neto_mean.append(neto_mean1)
    host1_neti.append(list(neti1)[59]/(1024**2))
    host1__neto.append(list(neto1)[59]/(1024**2))
#host2
    l_round_2 = vm2.query('round==%s'%i)
    cpu2 = l_round_2['cpu(%)']
    memoria2 = l_round_2['memoria(%)']
    host2_filtro.append(cpu2.tolist())
    cpu_mean2 = np.mean(cpu2)
    mem_mean2 = np.mean(memoria2)
    neti2 = l_round_2['Net-in(bytes)']
    neto2 = l_round_2['Net-out(bytes)']
    neti_mean2 = np.mean(neti2)
    neto_mean2 = np.mean(neto2)

    cpu_erro2 = st.t.interval(0.05, len(cpu2)-1, loc=np.mean(cpu2), scale=st.sem(cpu2))
    host2__cpu_erro.append(cpu_erro2[1] - cpu_erro2[0])

    mem_erro2 = st.t.interval(0.05, len(memoria2)-1, loc=np.mean(memoria2), scale=st.sem(memoria2))
    host2__mem_erro.append(mem_erro2[1] - mem_erro2[0])

    neti_erro2 = st.t.interval(0.05, len(neti2)-1, loc=np.mean(neti2), scale=st.sem(neti2))
    host2__neti_erro.append(neti_erro2[1] - neti_erro2[0])

    neto_erro2 = st.t.interval(0.05, len(neto2)-1, loc=np.mean(neto2), scale=st.sem(neto2))
    host2__neto_erro.append(neto_erro2[1] - neto_erro2[0])

    host2__cpu_mean.append(cpu_mean2)
    host2__mem_mean.append(mem_mean2)
    host2__neti_mean.append(neti_mean2)
    host2__neto_mean.append(neto_mean2)
    host2_neti.append(list(neti2)[59]/(1024**2))
    host2__neto.append(list(neto2)[59]/(1024**2))
#host3
    l_round_3 = vm3.query('round==%s'%i)
    cpu3 = l_round_3['cpu(%)']
    memoria3 = l_round_3['memoria(%)']
    host3_filtro.append(cpu3.tolist())
    cpu_mean3 = np.mean(cpu3)
    mem_mean3 = np.mean(memoria3)
    neti3 = l_round_3['Net-in(bytes)']
    neto3 = l_round_3['Net-out(bytes)']
    neti_mean3 = np.mean(neti3)
    neto_mean3 = np.mean(neto3)

    mem_erro3 = st.t.interval(0.05, len(memoria3)-1, loc=np.mean(memoria3), scale=st.sem(memoria3))
    host3__mem_erro.append(mem_erro3[1] - mem_erro3[0])

    neti_erro3 = st.t.interval(0.05, len(neti3)-1, loc=np.mean(neti3), scale=st.sem(neti2))
    host3__neti_erro.append(neti_erro3[1] - neti_erro3[0])

    neto_erro3 = st.t.interval(0.05, len(neto3)-1, loc=np.mean(neto3), scale=st.sem(neto3))
    host3__neto_erro.append(neto_erro3[1] - neto_erro3[0])

    cpu_erro3 = st.t.interval(0.05, len(cpu3)-1, loc=np.mean(cpu3), scale=st.sem(cpu3))
    host3__cpu_erro.append(cpu_erro3[1] - cpu_erro3[0])
    
    host3__cpu_mean.append(cpu_mean3)
    host3__mem_mean.append(mem_mean3)
    host3__neti_mean.append(neti_mean3)
    host3__neto_mean.append(neto_mean3)
    host3_neti.append(list(neti3)[59]/(1024**2))
    host3__neto.append(list(neto3)[59]/(1024**2))
'''
#host4
    l_round_4 = vm4.query('round==%s'%i)
    cpu4 = l_round_4['cpu(%)']
    memoria4 = l_round_4['memoria(%)']
    host4_filtro.append(cpu4.tolist())
    cpu_mean4 = np.mean(cpu4)
    mem_mean4 = np.mean(memoria4)
    neti4 = l_round_4['Net-in(bytes)']
    neto4 = l_round_4['Net-out(bytes)']
    neti_mean4 = np.mean(neti4)
    neto_mean4 = np.mean(neto4)

    mem_erro4 = st.t.interval(0.05, len(memoria4)-1, loc=np.mean(memoria4), scale=st.sem(memoria4))
    host4__mem_erro.append(mem_erro4[1] - mem_erro4[0])

    neti_erro4 = st.t.interval(0.05, len(neti4)-1, loc=np.mean(neti4), scale=st.sem(neti4))
    host4__neti_erro.append(neti_erro4[1] - neti_erro4[0])

    neto_erro4 = st.t.interval(0.05, len(neto4)-1, loc=np.mean(neto4), scale=st.sem(neto4))
    host4__neto_erro.append(neto_erro4[1] - neto_erro4[0])

    cpu_erro4 = st.t.interval(0.05, len(cpu3)-4, loc=np.mean(cpu4), scale=st.sem(cpu4))
    host4__cpu_erro.append(cpu_erro4[1] - cpu_erro4[0])

    host4__cpu_mean.append(cpu_mean4)
    host4__mem_mean.append(mem_mean4)
    host4__neti_mean.append(neti_mean4)
    host4__neto_mean.append(neto_mean4)
    host4_neti.append(list(neti4)[59]/(1024**2))
    host4__neto.append(list(neto4)[59]/(1024**2))
'''
```
